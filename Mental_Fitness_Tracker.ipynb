# Splitting the dataset

```python
X = data.drop('Mental_fitness', axis = 1)
y = data['Mental_fitness']
```

```python
# splitting the data
xtrain, xtest, ytrain, ytest = train_test_split(X,y, test_size = 0.20, random_state = 2)
```

```python
print('xtrain shape',xtrain.shape)
print('xtest shape',xtest.shape)

print('ytrain shape',ytrain.shape)
print('ytest shape',ytest.shape)
```

# Model Training

* **Linear Regression :**  Linear regression analysis is used to predict the value of a variable based on the value of another variable. The variable you want to predict is called the dependent variable. The variable you are using to predict the other variable's value is called the independent variable.

* **Random Forest**  is an ensemble technique capable of performing both regression and classification tasks with the use of multiple decision trees
 A random forest regressor. A random forest is a meta estimator that fits a number of classifying decision trees on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting.

* Linear Regression

```python
lr = LinearRegression()  # model creation
lr.fit(xtrain, ytrain)  # trainnig data fitting into model
```

```python
y_predict = lr.predict(xtrain) # predicting the test value
```

```python
# the mean square error is  the mean or average of the squared differences between the actual and estimated values
l_mse = mean_squared_error(ytrain,y_predict)

# Root mean squared error (RMSE) is the square root of the mean of the square of all of the error.
# Root Mean Square Error is the measure of how well a regression line fits the data points.
l_rmse = (np.sqrt(mean_squared_error(ytrain,y_predict)))

# r2 referred to as the coefficient of determination or the coefficient of multiple determination in case of multiple regression.
# R squared in regression acts as an evaluation metric to evaluate the scatter of the data points around the fitted regression line.
# You can use the R2 score to get the accuracy of your model on a percentage scale, that is 0100.
l_r2 = r2_score(ytrain, y_predict)

print('The performance of linear regression model for trainig set')
print('...........................')
print(f'Mean Squared Error is: {l_mse}\n')
print(f'Root Mean Squared Error is: {l_rmse}\n')
print(f'R square score is: {l_r2}')
```

* RandomForest Regressor

```python

# model creation
rf = RandomForestRegressor()

# model trainnig for training dataset
rf.fit(xtrain, ytrain)
```

```python
# model evaluation for training set
rf_y_predict = rf.predict(xtrain)
```

```python
# accuracy and error checking
r_mse = mean_squared_error(ytrain,rf_y_predict)
r_rmse = (np.sqrt(mean_squared_error(ytrain,rf_y_predict)))
r_r2 = r2_score(ytrain, rf_y_predict)

print('The performance of RandomForest Regressior model for trainig set')
print('...........................')
print(f'Mean Squared Error is: {r_mse}\n')
print(f'Root Mean Squared Error is: {r_rmse}\n')
print(f'R square score is: {r_r2}')
```

* Decision Tree regressor


```python
# model creation
d_model = DecisionTreeRegressor()

# model trainnig for training dataset
d_model.fit(xtrain, ytrain)
```

```python
# model evaluation for training set
d_y_predict = d_model.predict(xtrain)
```

```python
# accuracy and error checking
s_mse = mean_squared_error(ytrain,d_y_predict)
s_rmse = (np.sqrt(mean_squared_error(ytrain,d_y_predict)))
s_r2 = r2_score(ytrain, d_y_predict)

print('The performance of Decision Tree regressor model for trainig set')
print('...........................')
print(f'Mean Squared Error is: {s_mse}\n')
print(f'Root Mean Squared Error is: {s_rmse}\n')
print(f'R square score is: {s_r2}')
```

# Evaluation

* in this part we'll compare the score of two above models.

```python
# Linear Regression model evaluation for testing set
y_test_predict = lr.predict(xtest)

l_mse = mean_squared_error(ytest,y_test_predict)
l_rmse = (np.sqrt(mean_squared_error(ytest,y_test_predict)))
l_r2 = r2_score(ytest, y_test_predict)

print('The performance of linear regression model for testing set')
print('...........................')
print(f'Mean Squared Error is: {l_mse}')
print(f'Root Mean Squared Error is: {l_rmse}')
print(f'R square score is: {l_r2}')
print('__________________________________________\n\n')



# randomForest Regressor model evaluation for testing set
rf_ytest_predict = rf.predict(xtest)
r_mse = mean_squared_error(ytest,rf_ytest_predict)
r_rmse = (np.sqrt(mean_squared_error(ytest,rf_ytest_predict)))
r_r2 = r2_score(ytest, rf_ytest_predict)

print('The performance of RandomForest Regressior model for testing set')
print('...........................')
print(f'Mean Squared Error is: {r_mse}')
print(f'Root Mean Squared Error is: {r_rmse}')
print(f'R square score is: {r_r2}')
print('__________________________________________\n\n')


# model evaluation for testing set
d_y_predict = d_model.predict(xtest)
# accuracy and error checking
s_mse = mean_squared_error(ytest,d_y_predict)
s_rmse = (np.sqrt(mean_squared_error(ytest,d_y_predict)))
s_r2 = r2_score(ytest, d_y_predict)

print('The performance of Decision Tree regressor model for trainig set')
print('...........................')
print(f'Mean Squared Error is: {s_mse}')
print(f'Root Mean Squared Error is: {s_rmse}')
print(f'R square score is: {s_r2}')
print('__________________________________________')
```
